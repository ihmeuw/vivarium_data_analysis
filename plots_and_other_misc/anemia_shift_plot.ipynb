```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd, numpy as np
```

```python
plt.plot([0,0.5,2,5],[0,0,3,3])
plt.plot([1,1.5,2,5],[0,3*(1.5-0.5)/(1.5),3,3],'--')
plt.plot([2,2.5,5],[0,3,3],'--')
plt.plot([3,3.5,5],[0,3,3],'--')

plt.plot([0.5,0.5],[0,3], linestyle=(0, (1, 10)),c='gray')
plt.plot([2,2],[0,3], linestyle=(0, (1, 10)), c='gray')

plt.legend(['Covered at birth',
            'Coverage gained at one year',
            'Coverage gained at two years',
            'Coverage gained at three years'])

plt.title('Counterfactual Shift in Hemoglobin')
plt.xlabel('Age (years)')
plt.ylabel('Shift in Hemoglobin (g/L)')
plt.savefig('iron_scale_up.svg')
```
