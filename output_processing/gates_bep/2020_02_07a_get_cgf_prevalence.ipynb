```python
def get_cgf_prevalence_for_one_location(cgf_data):
    exposed_cats = ['cat1', 'cat2']
    index_cols = [col for col in cgf_data.columns if col not in ['measure', 'value']]
    exposed = cgf_data.query("measure in @exposed_cats")
    exposed = exposed.groupby(index_cols)['value'].sum()
    total = cgf_data.groupby(index_cols)['value'].sum()
    percent_exposed = (100*exposed/total).reset_index()
    percent_exposed['exposure_category'] = 'Z < -2'
    percent_exposed['metric'] = 'percent'
    return percent_exposed
```

```python
malnutrition_dir = '/share/costeffectiveness/results/vivarium_gates_bep/malnutrition/'
mali_folder = 'mali/2020_02_05_03_02_02/count_space/'
mali_cgf = pd.read_hdf(malnutrition_dir+mali_folder+'category_counts.hdf')
mali_cgf.head()
```

```python
def get_cgf_prevalence(cgf_data):
    exposed_cats = ['cat1', 'cat2']
    index_cols = [col for col in cgf_data.columns if col not in ['risk_category', 'category_counts']]
    exposed = cgf_data.query("risk_category in @exposed_cats")
    exposed = exposed.groupby(index_cols)['category_counts'].sum()
    total = cgf_data.groupby(index_cols)['category_counts'].sum()
    percent_exposed = (100*exposed/total).reset_index()
    percent_exposed.rename(columns={'category_counts': 'value'}, inplace=True)
    percent_exposed['risk_category'] = 'Z < -2'
    percent_exposed['metric'] = 'percent'
    return percent_exposed
```

```python
mali_cgf_prevalence = get_cgf_prevalence(mali_cgf.reset_index())
mali_cgf_prevalence.head()
```

```python
# The data was duplicated, so we'll drop half of it (and we don't want to stratify by sex anyway)
mali_cgf_prevalence = mali_cgf_prevalence.query("sex == 'female'")
mali_cgf_prevalence.drop(columns='sex', inplace=True)
mali_cgf_prevalence.head()
```
