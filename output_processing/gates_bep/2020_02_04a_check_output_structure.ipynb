## Check birth prevalence of NN disorders, as it looks like it might be wrong

In Mali, the artifact says birth prevalence of NN disorders should be ~0.21 for males and ~0.20 for females. In our sim it looks like it's ~0.087 for males and ~0.081 for females

```python
output.data.query("location == 'Mali'").head()
```

```python
nn_birth_prev_mali = output.data.query("location == 'Mali'").filter(regex='prevalent_count')
nn_birth_prev_mali = nn_birth_prev_mali.reset_index().groupby('input_draw_number').sum()
nn_birth_prev_mali.head()
```

```python
# Total population for each draw is 100,000
nn_birth_prev_mali = nn_birth_prev_mali.drop(columns='random_seed') / 100_000
nn_birth_prev_mali
```
