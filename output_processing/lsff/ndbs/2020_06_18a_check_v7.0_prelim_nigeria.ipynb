```python
# Filter out columns that are 0
live_births = output.subdata('live_births')
live_births = live_births.loc[:,(live_births>0).any()]
live_births
```

```python
# 1. Get all live births columns and scenario column
# 2. Rename scenario column 'scenario'
# 3. Add scenario column to index
live_births = output.subdata('live_births')
live_births = live_births.loc[:,(live_births>0).any()]
live_births = pd.concat([live_births, output.subdata('intervention')], axis=1)
live_births.rename(columns={'fortification_intervention.scenario': 'scenario'}, inplace=True)
live_births.set_index('scenario', append=True, inplace=True)
live_births
```

## 9. Check whether live births are equal across scenarios for each (draw, seed) pair

Add live births across all columns for each row, and sort the index to see if we get the same number in all four scenarios for each (draw, seed) pair.

Yes, we do. Woo hoo!

We can also check this by grouping by (draw, seed) and computing the variance -- all the variances should be 0.

```python
live_births.sum(axis=1).sort_index()
```

```python
live_births.sum(axis=1).groupby(level=['input_draw_number', 'random_seed']).var()
```

```python
def test_live_births(raw_output):
    live_births = raw_output.filter(regex='live_births')
    live_births = live_births.sum(axis=1)
    live_births_variance = live_births.groupby(level=live_births.index.names).var()
    assert np.allclose(live_births_variance, 0), \
        'live births should be equal across scenarios for each (draw, seed) pair'
    print('Live births test passed')
```

```python
test_live_births(raw_output)
```
