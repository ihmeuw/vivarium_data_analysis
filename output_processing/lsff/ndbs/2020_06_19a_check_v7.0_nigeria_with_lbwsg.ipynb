## 7. Check whether birthweight changes between scenarios

Mean birthweight should be larger in iron fortification scenario than in baseline, and it should be equal to baseline in other scenarios. In this output it is the same between all scenarios, which is incorrect.

The intention of the below function is to automate this check, but I haven't finished writing the function yet.

```python
output.columns('distribution_mean')
```

```python
bw_mean = raw_output.filter(regex=r'^birth_weight_mean$|scenario')
bw_mean.set_index('fortification_intervention.scenario', append=True, inplace=True)
bw_mean.sort_index()
```

```python
def test_birthweight_between_scenarios(raw_output):
    bw_mean = raw_output.filter(regex=r'^birth_weight_mean$|scenario')
    bw_mean = bw_mean.rename(columns={'fortification_intervention.scenario': 'scenario'})
    bw_mean_baseline = bw_mean.query('scenario == "baseline"')
    bw_mean_iron = bw_mean.query('scenario == "iron_fortification_scale_up"')
    bw_mean_not_iron = bw_mean.query('scenario != "iron_fortification_scale_up"')
#     bw_mean_not_iron.set_index('scenario', append=True, inplace=True)
    bw_mean_not_iron_var = bw_mean_not_iron.groupby(level=bw_mean_not_iron.index.names).var()
    assert np.allclose(bw_mean_not_iron_var, 0), \
        "mean birthweight should be equal in all scenarios except iron fortification"
     
    return bw_mean_baseline, bw_mean_iron
```

```python
df1, df2 = test_birthweight_between_scenarios(raw_output)
```
