# What is the distribution of $F_X(X)$ when $X$ is a discrete random variable and $F_X$ is its distribution function?

```python
p = 0.2
bernoulli = stats.bernoulli(p)

plotting.plot_rv_func(bernoulli, 'pmf')
plotting.plot_rv_func(bernoulli, 'cdf')
```

```python
bernoulli_rvs = bernoulli.rvs(10000)
plt.hist(bernoulli_rvs)
```

```python
plt.hist(bernoulli.cdf(bernoulli_rvs))
```

```python
p = 0.8
n = 50
binomial = stats.binom(n,p)
plotting.plot_rv_func(binomial, 'pmf', 0, n, n+1, False, linestyle='', marker='o')
plotting.plot_rv_func(binomial, 'cdf')
```

```python
binomial_rvs = binomial.rvs(1_000_000)
plt.hist(binomial_rvs, bins=3*n, density=True);
```

```python
plt.hist(binomial.cdf(binomial_rvs), bins=3*n, density=True);
```

```python
binomial.support()
```

```python
p = 0.4
geometric = stats.geom(p)
ub = int(geometric.ppf(0.999))
print(ub)
plotting.plot_rv_func(
    geometric, 'pmf', 1, ub, ub, False, linestyle='', marker='o', label='pmf')
plotting.plot_rv_func(geometric, 'cdf', label='cdf')
plt.legend()
```

```python
geometric_rvs = geometric.rvs(1_000_000)
plt.hist(geometric_rvs, bins=5*ub, density=True);
```

```python
plt.hist(geometric.cdf(geometric_rvs), bins=5*ub, density=True);
```

```python
mu = 3
poisson = stats.poisson(mu)
ub = int(poisson.ppf(0.999))
print(ub)
plotting.plot_rv_func(
    poisson, 'pmf', 0, ub, ub+1, False, linestyle='', marker='o', label='pmf')
plotting.plot_rv_func(poisson, 'cdf', label='cdf')
plt.legend()
```

```python
poisson_rvs = poisson.rvs(1_000_000)
plt.hist(poisson_rvs, bins=10*ub, density=True);
```

```python
plt.hist(poisson.cdf(poisson_rvs), bins=10*ub, density=True);
```
