```python
## nigeria
```

```python
# nigeria location_id = 214
# kano location_id = 25337
# lagos location_id = 25342

# from db_queries import get_population as go
# go(age_group_id = 22, location_id = [25337,25342], year_id = 2015, sex_id = 3, gbd_round_id = 7, decomp_step = 'iterative')

kano_pop_2015 = 1.389539e+07
lagos_pop_2015 = 7.192101e+06

kano_weight = kano_pop_2015/(kano_pop_2015+lagos_pop_2015)
lagos_weight = lagos_pop_2015/(kano_pop_2015+lagos_pop_2015)
```

```python
# pop eating ind prod vehicle
nigeria_ind_prod = nigeria.loc[(nigeria.value_description=="percent of population eating industrially produced vehicle") & 
     (nigeria.vehicle.isin(['wheat flour','oil'])),final_cols + ['subnational_name']]

nigeria_ind_prod.loc[nigeria_ind_prod.subnational_name=='lagos',['value_mean','value_025_percentile','value_975_percentile']] = nigeria_ind_prod.loc[nigeria_ind_prod.subnational_name=='lagos',['value_mean','value_025_percentile','value_975_percentile']] * lagos_weight

nigeria_ind_prod.loc[nigeria_ind_prod.subnational_name=='kano',['value_mean','value_025_percentile','value_975_percentile']] = nigeria_ind_prod.loc[nigeria_ind_prod.subnational_name=='kano',['value_mean','value_025_percentile','value_975_percentile']] * kano_weight

nigeria_ind_prod.groupby(['location_id',
 'location_name',
 'sub_population',
 'vehicle',
 'value_description',
 'nutrient']).sum().reset_index()[final_cols]
```

```python
output = output.append(nigeria_ind_prod.groupby(['location_id',
 'location_name',
 'sub_population',
 'vehicle',
 'value_description',
 'nutrient']).sum().reset_index()[final_cols])
```

```python
nigeria_fort = nigeria.loc[(nigeria.value_description=="percent of population eating fortified vehicle") & 
     (nigeria.vehicle.isin(['wheat flour','oil'])),final_cols + ['subnational_name']]
```

```python
nigeria_fort = nigeria.loc[(nigeria.value_description=="percent of population eating fortified vehicle") & 
     (nigeria.vehicle.isin(['wheat flour','oil']) & (nigeria.nutrient.isin(['folic acid','vitamin a','zinc','iron']))),final_cols + ['subnational_name']]
nigeria_fort
```

```python
nigeria_fort.loc[nigeria_fort.subnational_name=='lagos',['value_mean','value_025_percentile','value_975_percentile']] = nigeria_fort.loc[nigeria_fort.subnational_name=='lagos',['value_mean','value_025_percentile','value_975_percentile']] * lagos_weight

nigeria_fort.loc[nigeria_fort.subnational_name=='kano',['value_mean','value_025_percentile','value_975_percentile']] = nigeria_fort.loc[nigeria_fort.subnational_name=='kano',['value_mean','value_025_percentile','value_975_percentile']] * kano_weight

nigeria_fort.groupby(['location_id',
 'location_name',
 'sub_population',
 'vehicle',
 'value_description',
 'nutrient']).sum().reset_index()[final_cols]
```

```python
output = output.append(nigeria_fort.groupby(['location_id',
 'location_name',
 'sub_population',
 'vehicle',
 'value_description',
 'nutrient']).sum().reset_index()[final_cols])
```

```python
output.value_mean
```

```python
output.value_mean = output.value_mean.astype(float)
output.value_025_percentile = output.value_025_percentile.astype(float)
output.value_975_percentile = output.value_975_percentile.astype(float)
```

```python
# output.loc[output.value_025_percentile.isna(),'value_025_percentile'] = output.value_mean - 10
```

```python
output[(output.value_025_percentile.isna()) | (output.value_975_percentile.isna())]
```

```python
output.loc[output.value_025_percentile.isna(),'value_025_percentile'] = [i - 10 for i in output.loc[output.value_025_percentile.isna(),'value_mean']]
output.loc[output.value_975_percentile.isna(),'value_975_percentile'] = [i + 10 for i in output.loc[output.value_975_percentile.isna(),'value_mean']]
```
