```python
locations = [163, 214, 179] # india, nigeria, ethiopia
ages = [2,3,4,5]
```

```python
# get overall anemia impairment prevalence 

master_impairment = pd.DataFrame()

for location in locations:
    for severity in [192,205,206,207]:
        anemia = get_draws('rei_id', severity, source='como', measure_id=5, 
                           location_id=location, year_id=2017, gbd_round_id=5)
        anemia_all = anemia.where(anemia['cause_id'] == 294).dropna()
        anemia_avg = anemia_all.set_index(['age_group_id','location_id', 'measure_id', 'rei_id',
                                         'sex_id', 'year_id', 'metric_id','cause_id'])
        anemia_avg['avg'] = anemia_avg.mean(axis=1)
        anemia_avg = anemia_avg.reset_index()
        data = anemia_avg[['age_group_id','sex_id','avg']]
        data = data.rename(columns={'avg': 'prev'})
        data = data.where(data['age_group_id'].isin([2,3,4,5])).dropna()
        data['location_id'] = location
        data['severity_id'] = severity
        master_impairment = master_impairment.append(data)
        
master_impairment.head()       
```

```python
impair = master_impairment#.drop(columns='severity_id')
impair['severity'] = np.where(impair['severity_id'] == 192, 'overall',
                             np.where(impair['severity_id'] == 205, 'mild',
                                     np.where(impair['severity_id'] == 206, 'moderate','severe')))
impair.head()
```

```python
# get total anemia prevalence summed across all causes

severity_list = [mild_all, moderate_all, severe_all, overall_all]
name_list = ['mild','moderate','severe','overall']
master_summed_total = pd.DataFrame()

for i in list(range(len(severity_list))):
    data = get_outputs(gbd_round_id=5, year_id=2017, location_id=locations, age_group_id=ages,
                       sex_id=[1,2],
                      topic='sequela', sequela_id=severity_list[i], measure_id=5, metric_id=3)
    data = data.groupby(['location_id','age_group_id','sex_id'], as_index=False).sum()
    data = data[['location_id','sex_id','age_group_id','val']]
    data['severity'] = name_list[i]
    master_summed_total = master_summed_total.append(data)

master_summed_total = master_summed_total.rename(columns={'val':'sum_tot'})
master_summed_total.head()
```

```python
# get iron responsive anemia prevalence summed across all causes

severity_list = [mild_ira, moderate_ira, severe_ira, overall_ira]
name_list = ['mild','moderate','severe','overall']
master_summed_ira = pd.DataFrame()

for i in list(range(len(severity_list))):
    data = get_outputs(gbd_round_id=5, year_id=2017, location_id=locations, age_group_id=ages,
                       sex_id=[1,2],
                      topic='sequela', sequela_id=severity_list[i], measure_id=5, metric_id=3)
    data = data.groupby(['location_id','age_group_id','sex_id'], as_index=False).sum()
    data = data[['location_id','sex_id','age_group_id','val']]
    data['severity'] = name_list[i]
    master_summed_ira = master_summed_ira.append(data)
    
master_summed_ira = master_summed_ira.rename(columns={'val':'sum_ira'})
master_summed_ira.head()
```

```python
test = (master_summed_total.merge(impair, 
                                right_on=['location_id','sex_id','age_group_id','severity'],
                                left_on=['location_id','sex_id','age_group_id','severity'])
        .drop(columns='severity_id'))
test = test.merge(master_summed_ira, 
                                right_on=['location_id','sex_id','age_group_id','severity'],
                                left_on=['location_id','sex_id','age_group_id','severity'])
test['diff'] = test['sum_tot'] - test['prev']
test['pct'] = test['sum_ira'] / test['prev']
test.head()
```

```python
test['diff'].max() # anemia impairment prevalence ~ sum across all anemia sequelae
```

```python
# plot probability of iron-responsive anemia in each location of interest
# can change sex and anemia severity of interest in first line of this code

data = test.where(test['sex_id'] == 2).where(test['severity'] == 'overall').dropna()

barwidth = 0.2

early = data['pct'].where(data['age_group_id'] == 2).dropna()
late = data['pct'].where(data['age_group_id'] == 3).dropna()
post = data['pct'].where(data['age_group_id'] == 4).dropna()
one = data['pct'].where(data['age_group_id'] == 5).dropna()

# bar positions
a = np.arange(len(early)) 
b = [x + barwidth for x in a]
c = [x + barwidth for x in b]
d = [x + barwidth for x in c]

plt.bar(a, early, width=barwidth)
plt.bar(b, late, width=barwidth)
plt.bar(c, post, width=barwidth)
plt.bar(d, one, width=barwidth)

plt.title('Probability of Iron Responsive Anemia - Boys')
plt.legend(['Early NN','Late NN','Post NN','1-4 y'])
plt.xticks([r + barwidth for r in range(len(early))], ['India','Ethiopia','Nigeria'])
plt.ylim(0,1.4)
```
